```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 


fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

df1 = [[1404,2],[50,0]]
df2 = [[284030,285],[487,5]]
df3 = [[5060,6],[150,0]]
sns.heatmap(df1, annot = True, fmt = 'd',cmap = 'Pastel1',ax=ax1)
sns.heatmap(df2, annot = True, fmt = 'd',cmap = 'Pastel1',ax=ax2)
sns.heatmap(df3, annot = True, fmt = 'd',cmap = 'Pastel1',ax=ax3)

ax1.set_title('data_1')
ax2.set_title('data_2')
ax3.set_title('data_3')

plt.savefig('./pics/IF_3D.png')
```
